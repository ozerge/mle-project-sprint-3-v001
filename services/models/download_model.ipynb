```python
import boto3
import os
import pandas as pd
import joblib
from dotenv import load_dotenv, find_dotenv
import psycopg2 as psycopg
load_dotenv()
```

```python
def get_s3_client():
    return boto3.client(
        service_name='s3',
        endpoint_url='https://storage.yandexcloud.net',
        aws_access_key_id=os.environ.get("AWS_ACCESS_KEY_ID"),
        aws_secret_access_key=os.environ.get("AWS_SECRET_ACCESS_KEY")
    )

s3 = get_s3_client()
bucket_name = os.environ.get('S3_BUCKET_NAME')

# Получаем и сортируем объекты по дате создания
objects = s3.list_objects(Bucket=bucket_name).get('Contents', [])
sorted_objects = sorted(objects, key=lambda x: x['LastModified'])

if sorted_objects:
    print("Объекты в бакете (отсортированы по дате создания):")
    for obj in sorted_objects:
        print(obj['Key'])
else:
    print("Бакет пуст.")
```

```python
# Настройки AWS
AWS_CONFIG = {
    'aws_access_key_id': os.environ.get("AWS_ACCESS_KEY_ID"),
    'aws_secret_access_key': os.environ.get("AWS_SECRET_ACCESS_KEY"),
    'endpoint_url': "https://storage.yandexcloud.net",
    'region_name': "ru-central1"
}
BUCKET_NAME = os.environ.get('S3_BUCKET_NAME')

# Конфигурация файлов для загрузки
FILES_TO_DOWNLOAD = [
    {
        's3_path': "models/fitted_model_cbr_l.pkl",
        'local_filename': "fitted_model_cbr_l.pkl"  # или можно просто взять имя файла из s3_path
    },
]

# Инициализация S3 клиента
s3 = boto3.client('s3', **AWS_CONFIG)

# Загрузка файлов
for file_config in FILES_TO_DOWNLOAD:
    # Формируем путь к файлу в текущей директории
    local_path = file_config['local_filename']
    
    # Скачиваем файл
    s3.download_file(BUCKET_NAME, file_config['s3_path'], local_path)
    print(f"Файл скачан из S3 в {os.path.abspath(local_path)}")
```

```python
# Загружаем сохранённый pipeline
pipeline = joblib.load('fitted_model_cbr_l.pkl')

# Просмотр сохранённых метаданных
print("Feature names:", pipeline.feature_names)
```
